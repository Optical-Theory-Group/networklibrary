```python
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch
```

```python
fig, ax = plt.subplots()

x1 = 0.0
y1 = 0.0
x2 = 1.0
y2 = 5.0
ax.plot((x1,x2), (y1,y2))

cx = (x1+x2) / 2
cy = (y1+y2) / 2

arrow = FancyArrowPatch(
    (cx, cy),  # Starting point of the arrowhead
    (cx + 0.1, cy + 0.5),  # Ending point of the arrowhead
    arrowstyle='->',  # Arrow style
    mutation_scale=200,  # Controls the size of the arrowhead
    color='blue',  # Arrow color
)

# Add the arrow to the axis
ax.add_patch(arrow)
```
