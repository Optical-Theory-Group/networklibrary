```python
# Lets do some ipr tests
import os
import sys
import h5py
import numpy as np
```

```python
# Open the hdf5 file
with h5py.File(r'/home/baruva/Sacttering_Data/scattering_ensemble_del_circ_30_10.h5', 'r') as hdfile:
    length = len(hdfile['S_ee'])
    IPR = []
    for i in range(0,length,1):
        S_ee = hdfile['S_ee'][f'matrix_idx={i}'][:]
        
        # Lets find the eigen values and eigen vectors
        eigvals, eigvecs = np.linalg.eig(S_ee)

        # print(eigvecs)
        IPR = IPR+ list(np.sum(np.pow(np.abs(eigvecs),4),axis=0))

# Open the hdf5 file
with h5py.File(r'/home/baruva/Sacttering_Data/scattering_ensemble_del_circ_50_10.h5', 'r') as hdfile:
    length = len(hdfile['S_ee'])
    IPR2 = []
    for i in range(0,length,1):
        S_ee = hdfile['S_ee'][f'matrix_idx={i}'][:]
        
        # Lets find the eigen values and eigen vectors
        eigvals, eigvecs = np.linalg.eig(S_ee)

        # print(eigvecs)
        IPR2 = IPR2+ list(np.sum(np.pow(np.abs(eigvecs),4),axis=0))

# Open the hdf5 file
with h5py.File(r'/home/baruva/Sacttering_Data/scattering_ensemble_del_circ_100_10.h5', 'r') as hdfile:
    length = len(hdfile['S_ee'])
    IPR3 = []
    for i in range(0,length,1):
        S_ee = hdfile['S_ee'][f'matrix_idx={i}'][:]
        
        # Lets find the eigen values and eigen vectors
        eigvals, eigvecs = np.linalg.eig(S_ee)

        # print(eigvecs)
        IPR3 = IPR3 + list(np.sum(np.pow(np.abs(eigvecs),4),axis=0))

# Open the hdf5 file
with h5py.File(r'/home/baruva/Sacttering_Data/scattering_ensemble_del_circ_500_10.h5', 'r') as hdfile:
    length = len(hdfile['S_ee'])
    IPR4 = []
    for i in range(0,288,1):
        S_ee = hdfile['S_ee'][f'matrix_idx={i}'][:]
        
        # Lets find the eigen values and eigen vectors
        eigvals, eigvecs = np.linalg.eig(S_ee)

        # print(eigvecs)
        IPR4 = IPR4 + list(np.sum(np.pow(np.abs(eigvecs),4),axis=0))
```
