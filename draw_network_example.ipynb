```python
import matplotlib.pyplot as plt
import numpy as np

from complex_network.networks import network_factory
from complex_network.networks.network_spec import NetworkSpec
%matplotlib widget
```

```python
type_cases = [
    "delaunay",
    "voronoi",
    "buffon",
]
shape_cases = [
    "circular",
    "slab",
]

np.random.seed(5)
plt.figure() 
plot_number = 0
for network_shape in shape_cases:
    for network_type in type_cases:
        print(f"Generating {network_type} {network_shape} network")
        plot_number += 1
        spec = NetworkSpec(
            network_type=network_type,
            network_shape=network_shape,
        )

        network = network_factory.generate_network(spec)

        plt.subplot( len(shape_cases), len(type_cases), plot_number)
        network.draw(
            equal_aspect=True,
            highlight_links=[],
            show_external_indices=False,
            show_indices=False,
            highlight_nodes=[],
            draw_boundary = spec.network_size 
        )
        plt.title(f"{network_type} {network_shape}")

```
