# ðŸŽ¨ Random network examples
This Jupyter notebook generates and visualizes several example random networks of different shapes and types.

```python
import matplotlib.pyplot as plt
import numpy as np

from complex_network.networks import network_factory
from complex_network.networks.network_spec import NetworkSpec
```

Note that networks should be generated using methods from the `network_factory.py` module, located in the `complex_network/networks/` directory. These methods take a `NetworkSpec` object, which is defined in `network_spec.py` within the same directory. This approach is preferred over directly initializing a `Network` object.

```python
np.random.seed(5)

type_cases = [
    "delaunay",
    "voronoi",
    "buffon",
]
shape_cases = [
    "circular",
    "slab",
]

fig, ax = plt.subplots(len(shape_cases), len(type_cases))
for i, network_shape in enumerate(shape_cases):
    for j, network_type in enumerate(type_cases):
        spec = NetworkSpec(
            network_type=network_type,
            network_shape=network_shape,
        )
        network = network_factory.generate_network(spec)
        network.draw(
            ax=ax[i, j],
            equal_aspect=True,
            highlight_links=[],
            show_external_indices=False,
            show_indices=False,
            highlight_nodes=[],
            draw_boundary=spec.network_size,
        )
        ax[i, j].set_title(f"{network_type} {network_shape}")
```
